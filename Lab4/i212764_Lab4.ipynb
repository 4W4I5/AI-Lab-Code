## Online Store Shopping Cart
Write a Python program to simulate the online store's shopping cart system. The program should use a while loop to provide an interactive shopping experience.


### Tasks:
* Initialize an empty list called shopping_cart to store items.
* Display a welcome message and available products.
* Loop should be iterating over AI Response Simulator and Investment decision and keep asking for user input.
* Prompt the user to enter the number corresponding to the product they want to add to the cart.
* Check if the user entered 'q'. If so, break the loop.
* Convert the choice to an integer and validate it within the valid range (1 to 3).
* Based on the user's choice, add the selected product to the shopping_cart list.
* Display a confirmation message indicating the added product.
* Continue the loop to allow the user to add more products or proceed to checkout.
* After the loop ends, display the contents of the shopping cart.
* Display a goodbye message.
* End the program.


```python
shopping_cart = []

print("Welcome to the Online Store!")
print("Available Products:")
print("1. Product A")
print("2. Product B")
print("3. Product C\n")

# Loop to provide an interactive shopping experience
while True:
    choice = input("Enter the number of the product you want to add to the cart (1-3), or 'q' to quit: ")
    
    # Check if the user entered 'q'. If so, break the loop
    if choice == 'q':
        break
    
    # Convert the choice to an integer and validate it within the valid range (1 to 3)
    choice = int(choice)
    if choice < 1 or choice > 3:
        print("Invalid choice. Please enter a number between 1 and 3.")
        continue
    
    # add the selected product to the shopping_cart list
    if choice == 1:
        shopping_cart.append("Product A")
    elif choice == 2:
        shopping_cart.append("Product B")
    elif choice == 3:
        shopping_cart.append("Product C")
    
    print("Product added to the shopping cart.")
    
print("\nShopping Cart:")
for item in shopping_cart:
    print(item)
    
print("Thank you for shopping with us!")

```

## Email Classification
Develope a tool which will classify an email.

### Tasks:
* Create a function classify_emails that takes a list of emails as an argument.
* Inside the function, categorize emails into "Important," "Promotions," and "Spam" based on keywords and content analysis.
* Loop through the emails and identify keywords to determine their category.
* If an email contains keywords indicating importance, assign it as "Important."
* If an email contains keywords related to promotions or deals, assign it as "Promotions."
* If an email is suspected to be spam based on certain keywords or patterns, assign it as "Spam."
* Return three lists of emails for each category.
* Print the lists.

```python
def classify_emails(emails, keywords):
    
    categorized_emails = {category: [] for category in keywords}
    
    for email in emails:
        email_lower = email.lower()
        
        for category, category_keywords in keywords.items():
            for keyword in category_keywords:
                if keyword in email_lower:
                    categorized_emails[category].append(email)
                    break
    
    return categorized_emails


emails = [
    "Hello, this is an urgent email.",
    "Check out our latest sale!",
    "You have won the lottery!",
    "Important: Please review the attached document.",
    "Get a 50% discount on our products.",
    "Congratulations! You have been selected as a winner.",
    "Limited time offer: Buy one, get one free!",
    "Make money fast with our investment opportunity.",
    "Important notice: Update your account information.",
    "Claim your prize now!",
    "Click on this link"
]

keywords = {
        'important': ['urgent', 'important', 'priority'],
        'promotions': ['sale', 'discount', 'offer'],
        'spam': ['link', 'lottery', 'inheritance']
    }
categorized_emails = classify_emails(emails, keywords)

print("\nAll Emails")
for category, emails in categorized_emails.items():
    x = 1
    print(f"{category.capitalize()}:")
    for email in emails:
        print(f"{x}) {email}")
        x+=1
    print()   
```

#Exercise
Use the dataset of Game,  perform following operations.

* Calculate the total global sales for all games in the dataset.
* Find the game with the highest global sales.
* Determine the average sales for each region (NA_Sales, EU_Sales, JP_Sales, Other_Sales).
* Find the top 5 most popular game genres based on global sales.
* Calculate the total sales for each genre in each region and find the most popular genre in North America (NA_Sales).
* Display last 10 rows.
* Determine the top 3 gaming platforms with the highest global sales.
* Calculate the market share (percentage of global sales) for each genre.
* Calculate the correlation coefficient between NA_Sales and EU_Sales. (.corr function)

```python
import pandas as pd

df = pd.read_csv("vgsalesGlobale.csv")

# Global Sum for every games
total_global_sales = df["Global_Sales"].sum()
print(f"Total Global Sales: {total_global_sales}\n")

# Find the game with the highest global sales
game_with_highest_sales = df.loc[df["Global_Sales"].idxmax(), "Name"]
print(f"Game with Highest Global Sales: \n{game_with_highest_sales}\n")

# Determine the average sales for each region
average_sales = df[["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales"]].mean()
print(f"Average Sales for Each Region: \n{average_sales}\n")

# Find the top 5 most popular game genres based on global sales
top_5_genres = df.groupby("Genre")["Global_Sales"].sum().nlargest(5)
print(f"Top 5 Most Popular Game Genres: \n{top_5_genres}\n")

# Calculate the total sales for each genre in each region and find the most popular genre in North America
genre_sales_by_region = df.groupby(["Genre"]).agg({"NA_Sales": "sum", "EU_Sales": "sum", "JP_Sales": "sum", "Other_Sales": "sum"})
most_popular_genre_na = genre_sales_by_region["NA_Sales"].idxmax()
print(f"Most Popular Genre in North America: \n{most_popular_genre_na}\n")

# Display last 10 rows
print(f"Last 10 Rows: \n{df.tail(10)}\n")

# Determine the top 3 gaming platforms with the highest global sales
top_3_platforms = df.groupby("Platform")["Global_Sales"].sum().nlargest(3)
print(f"Top 3 Gaming Platforms with Highest Global Sales: \n{top_3_platforms}\n")


# Calculate the market share (percentage of global sales) for each genre
genre_market_share = df.groupby("Genre")["Global_Sales"].sum() / df["Global_Sales"].sum() * 100
print(f"Market Share for Each Genre: \n{genre_market_share}\n")

# Calculate the correlation coefficient between NA_Sales and EU_Sales
correlation_coefficient = df["NA_Sales"].corr(df["EU_Sales"])
print(f"Correlation Coefficient between NA_Sales and EU_Sales: \n{correlation_coefficient}\n")
```
